### Cleaning World Data

```python
# Step 1: Convert Timestamp to datetime format
df = world_df.copy()

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Step 2: Handling Missing Values
# Drop 'comments' column due to high missing values
df.drop(columns=['comments'], inplace=True)

# Fill missing values in 'self_employed' with 'No' (assuming most are not self-employed)
df['self_employed'].fillna('No', inplace=True)

# Fill missing values in 'work_interfere' with 'Don't know' (assuming respondents didn't answer due to uncertainty)
df['work_interfere'].fillna("Don't know", inplace=True)

# Step 3: Handle 'state' missing values (Keep it for now as it might be useful in country analysis)
df['state'].fillna('Unknown', inplace=True)

# Display updated missing values
df.isnull().sum()
```

```python
# Step 4: Clean and Standardize Gender Column

# Convert to lowercase and strip spaces
df['Gender'] = df['Gender'].str.lower().str.strip()

# Replace different variations of gender responses with standardized values
gender_mappings = {
    'male': ['male', 'm', 'man', 'male-ish', 'maile', 'mal', 'male (cis)', 'make', 'cis male'],
    'female': ['female', 'f', 'woman', 'femake', 'female (cis)', 'cis female'],
    'non-binary': ['non-binary', 'nb', 'genderqueer', 'androgyne', 'queer'],
    'other': ['trans-female', 'trans woman', 'trans male', 'trans man', 'agender', 'neither']
}

# Function to map gender values
def standardize_gender(gender):
    for key, values in gender_mappings.items():
        if gender in values:
            return key
    return 'other'  # Default to 'other' if it doesn't fit in main categories

# Apply standardization
df['Gender'] = df['Gender'].apply(standardize_gender)

# Display unique values after cleaning
df['Gender'].value_counts()
```

```python
import pandas as pd
import pandasql as psql

# Load the one-hot encoded dataset
file_path = "/content/cleaned_survey_onehot.csv"
df = pd.read_csv(file_path)

# Ensure consistent column names for SQL queries
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Define function to run SQL queries with pandasql
def run_query(query):
    return psql.sqldf(query, globals())  # Using global() to ensure df is accessible

# ------------------ UPDATED DATA CLEANING QUERIES ------------------

# 1️⃣ Fix Age Outliers & Categorize Age Groups
query_fix_age = """
    SELECT
        CASE
            WHEN age BETWEEN 18 AND 24 THEN '18-24'
            WHEN age BETWEEN 25 AND 34 THEN '25-34'
            WHEN age BETWEEN 35 AND 44 THEN '35-44'
            WHEN age BETWEEN 45 AND 54 THEN '45-54'
            WHEN age BETWEEN 55 AND 65 THEN '55+'
            ELSE 'Unknown'
        END AS age_group,
        COUNT(*) AS count
    FROM df
    WHERE age BETWEEN 18 AND 65  -- Removing extreme outliers
    GROUP BY age_group
    ORDER BY age_group;
"""

# 2️⃣ Standardize Gender Categories
query_fix_gender = """
    SELECT
        CASE
            WHEN gender = 'male' THEN 'Male'
            WHEN gender = 'female' THEN 'Female'
            ELSE 'Non-Binary/Other'
        END AS gender_category,
        COUNT(*) AS count
    FROM df
    GROUP BY gender_category
    ORDER BY count DESC;
"""

# ------------------ EXECUTE UPDATED CLEANING QUERIES ------------------

cleaning_queries = {
    "Age Group Distribution (Fixed)": query_fix_age,
    "Gender Standardization": query_fix_gender,
}

query_results_cleaned = {key: run_query(query) for key, query in cleaning_queries.items()}

for key, result in query_results_cleaned.items():
    print(f"\n--- {key} ---\n")
    print(result)
```
